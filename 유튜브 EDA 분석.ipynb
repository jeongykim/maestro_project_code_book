```python
def Plot_corr(data1):
    f,ax=plt.subplots(3,2,figsize=(20,10))
    
    sns.scatterplot(x='view_count',y='likes', data=data1, ax=ax[0][0])
    sns.scatterplot(x='view_count',y='dislikes', data=data1, ax=ax[0][1])
    sns.scatterplot(x='comment_count',y='likes', data=data1, ax=ax[1][0])
    sns.scatterplot(x='comment_count',y='dislikes', data=data1, ax=ax[1][1])
    sns.scatterplot(x='likes',y='dislikes', data=data1, ax=ax[2][0])
    
    

Plot_corr(data1)
```

### 1. 높은 상관도(Highly Correlated) : x=조회수/ y=좋아요, x=댓글 수/ y=좋아요
### 2. 보통의 상관도(Moderately Correlated) : 없음
### 3. 낮은 상관도(Lowly Correlated) : x=조회수/ y=싫어요, x=댓글 수/ y=싫어요, x=좋아요/ y=싫어요

```python
def Plot_hist(df,limit=None,factor='views', per_limit=None):
    if per_limit is not None:
        per=(len(df[df[factor]<per_limit][factor])/len(df[factor]) )*100
        print('{0} percent of trended videos got less than {1} {2}'.format(per,per_limit,factor))
    if limit is None:
        plt.figure(figsize=(14,7))
        sns.distplot(df[factor], kde=False, color='#976393')
        plt.ylabel('No. of videos')
    else:
        f,ax=plt.subplots(1,2 ,figsize=(20,7))
        ax[0].set( ylabel="No. of Videos")
        ax[1].set(ylabel="No. of Videos")
        sns.distplot(df[factor], kde=False, color='black', ax=ax[0])
        sns.distplot(df[df[factor] < limit][factor],kde=False, color='b', ax=ax[1])
```

```python
#불용어 설정
eng_stopwords = set(stopwords.words("english"))
```

```python
df['count_word']=df["title"].apply(lambda x: len(str(x).split()))
df['count_word_tags']=df["tags"].apply(lambda x: len(str(x).split()))

#고유한 단어 수
df['count_unique_word']=df["title"].apply(lambda x: len(set(str(x).split())))
df['count_unique_word_tags']=df["tags"].apply(lambda x: len(set(str(x).split())))

#문자 수
df['count_letters']=df["title"].apply(lambda x: len(str(x)))
df['count_letters_tags']=df["tags"].apply(lambda x: len(str(x)))

#구두점 수
df["count_punctuations"] =df["title"].apply(lambda x: len([c for c in str(x) if c in string.punctuation]))
df["count_punctuations_tags"] =df["tags"].apply(lambda x: len([c for c in str(x) if c in string.punctuation]))

#대문자 단어 수
df["count_words_upper"] = df["title"].apply(lambda x: len([w for w in str(x).split() if w.isupper()]))
df["count_words_upper_tags"] = df["tags"].apply(lambda x: len([w for w in str(x).split() if w.isupper()]))

#제목 케이스 단어 수
df["count_words_title"] = df["title"].apply(lambda x: len([w for w in str(x).split() if w.istitle()]))
df["count_words_title_tags"] = df["tags"].apply(lambda x: len([w for w in str(x).split() if w.istitle()]))

#불용어의 수
df["count_stopwords"] = df["title"].apply(lambda x: len([w for w in str(x).lower().split() if w in eng_stopwords]))
df["count_stopwords_tags"] = df["tags"].apply(lambda x: len([w for w in str(x).lower().split() if w in eng_stopwords]))

#단어의 평균 길이
df["mean_word_len"] = df["title"].apply(lambda x: np.mean([len(w) for w in str(x).split()]))
df["mean_word_len_tags"] = df["tags"].apply(lambda x: np.mean([len(w) for w in str(x).split()]))
```

```python
#파생 피쳐
#각 댓글의 단어 수 비율:
df['word_unique_percent']=df['count_unique_word']*100/df['count_word']
df['word_unique_percent_tags']=df['count_unique_word_tags']*100/df['count_word_tags']

#각 댓글의 백분율:
df['punct_percent']=df['count_punctuations']*100/df['count_word']
df['punct_percent_tags']=df['count_punctuations_tags']*100/df['count_word_tags']
```

```python
def Top_Channels(data1):
    Content_list=[24, 22, 10, 17]
    f,ax=plt.subplots(2, 2, figsize=(25,12))
    color=[sns.light_palette((210, 90, 60), input="husl",reverse=True), sns.cubehelix_palette(5,reverse=True),
          sns.light_palette("navy", reverse=True), sns.light_palette("green",reverse=True)]
    for i, content in enumerate(Content_list):
        ax[i//2][i%2].set(xlabel="No. of Videos Trended", ylabel="Channel")
        sns.barplot(x=data1[data1.categoryId==content].channelTitle.value_counts()[:5].values,
                    y=data1[data1.categoryId==content].channelTitle.value_counts()[:5].index,
                    ax=ax[i//2][i%2], palette=color[i] )
Top_Channels(data1)
```
