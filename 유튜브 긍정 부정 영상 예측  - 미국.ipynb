```python
import pandas as pd
data = pd.read_csv('USvideos.csv')
data.columns
```

```python
data['time'] = data.publish_time.str.slice(11,19)
```

```python
data.shape
```

```python
data
```

# 조회수 0회인 열 제외 (영상이 제거 되면 조회수가 0으로 기록)
# 트랜딩 업데이트가 되어 중복되는 title 제거 

```python
data2 = data[data['views']!=0]
data3 = data2.drop_duplicates("title")
```

```python
data3.shape
```

```python
data3.columns
```

```python
data3 = data3[['views','likes', 'dislikes', 'comment_count','tags','time','category_id','comments_disabled','ratings_disabled']]
```

```python
data3
```

# 댓글 막기, 평가 내린 것 안보이게 막기 범주 0,1 변환

```python
data3['comments_block'] = data3.comments_disabled.replace({False, True},{0, 1})
#data3.comments_disabled.replace(True, 1)
```

```python
data3['ratings_block'] = data3.ratings_disabled.replace({False, True},{0, 1})
```

# 출간 시간대 변수 변환

```python
data3.time
```

```python
data3['pub_time'] = pd.to_numeric(data3.time.str.slice(0,2))
```

```python
data3.dtypes # 문자열 파일 숫자형으로 변환 완료
```

# 태그 유무 전처리

```python
data4 = data3[data3.tags!='[none]']
```

```python
data4['tag_yn'] = 1
```

```python
data4
```

```python
data5 = data3[data3.tags=='[none]']
```

```python
data5['tag_yn'] = 0
```

```python
data5
```

```python
data6 = pd.concat([data4, data5])
```

```python
data6
```

# 만들고 필요한 변수들만 가져오기 1

```python
data6 = data6[['views','likes','dislikes','comment_count','category_id','comments_block','ratings_block','pub_time','tag_yn']]
```

```python
data6.columns
```

```python
data6 = data6[data6.likes!=0]
data6 = data6[data6.dislikes!=0]
```

```python
data6.shape
```

# 긍정 <= 부정 타겟 생성  부정이 높거나 같으면 1 낮으면 긍정

```python
data6['rate']= data6.likes/data6.dislikes
```

```python
data6.rate.describe()
```

# 긍정 보통 부정 target 선정

```python
data7 = data6[data6.rate>=67]
```

```python
data7['Target'] = '긍정'
```

```python
data8 = data6[(data6.rate < 67) & (data6.rate>=3.5)]
```

```python
data8['Target'] = '보통'
```

```python
data9 = data6[data6.rate < 3.5]
```

```python
data9['Target']='부정'
```

```python
data7.shape,data8.shape,data9.shape
```

```python
data7.shape[0]+data8.shape[0]+data9.shape[0]
```

```python
data10 = pd.concat([data7, data8, data9])
```

```python
data10
```

```python
data6.shape
```

```python
data10.Target.value_counts()
```

```python
data10.shape
```

```python
data10 = data10.rename(columns={'views':'view_count','category_id':'categoryId'})
```

```python
df = data10[['view_count','comment_count','categoryId','comments_block','ratings_block','pub_time','tag_yn','Target']]
```

```python
df
```

```python
from sklearn.tree import DecisionTreeClassifier
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
X_train, X_test, y_train,y_test = train_test_split(df.iloc[:,0:7], df.iloc[:,7:8], random_state=0)
```

```python
X_train.shape,X_test.shape
```

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression(max_iter=1000)
lr.fit(X_train, y_train)
lr.score(X_train, y_train),lr.score(X_test, y_test)
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=229)
knn.fit(X_train,y_train)
knn.score(X_train, y_train), knn.score(X_test, y_test)
```

```python
from sklearn.tree import DecisionTreeClassifier
tree = DecisionTreeClassifier(random_state=283)
tree.fit(X_train, y_train)
tree.score(X_train, y_train) , tree.score(X_test, y_test)
```

```python
from sklearn.ensemble import RandomForestClassifier
forest = RandomForestClassifier(n_estimators=70,random_state=3)
forest.fit(X_train,y_train)
forest.score(X_train, y_train) , forest.score(X_test, y_test)
```
